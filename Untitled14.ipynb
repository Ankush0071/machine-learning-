# Outlier Removal using IQR

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('loan_sanction_test.csv')
dataset.head(3)
```

```python
dataset.describe()
```

```python
sns.boxplot(x="CoapplicantIncome",data=dataset)
plt.show()
```

```python
dataset.shape
```

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)
```

```python
IQR = q3-q1
```

```python
min_range = q1-1.5*IQR
max_range = q3+1.5*IQR
min_range,max_range
```

```python
new_dataset = dataset[dataset["CoapplicantIncome"]<=max_range]
```

```python
new_dataset.shape
```

```python
sns.boxplot(x="CoapplicantIncome",data=new_dataset)
plt.show()
```

# Outlier Removal Using Z Score

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset =pd.read_csv("loan_sanction_test.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
sns.boxplot(x="CoapplicantIncome",data=dataset)
plt.show()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
min_range = dataset["CoapplicantIncome"].mean() - (3*dataset["CoapplicantIncome"].std()) # using formula Z score
max_range = dataset["CoapplicantIncome"].mean() + (3*dataset["CoapplicantIncome"].std())
min_range,max_range
```

```python
new_dataset=dataset[dataset["CoapplicantIncome"]<=max_range]
new_dataset.shape
```

```python
sns.boxplot(x="CoapplicantIncome",data=new_dataset)
plt.show()
```

```python
z_score = (dataset["CoapplicantIncome"] - dataset["CoapplicantIncome"].mean())/dataset["CoapplicantIncome"].std() #direct formula
z_score>3
```

```python
dataset["z_score"] = z_score
dataset
```

```python
dataset[dataset["z_score"]<3]
```

# Function Transformer

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("loan_sanction_test.csv")
dataset.head(3)
```

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)
IQR = q3-q1
```

```python
min_r = q1-(1.5*IQR)
max_r = q3+(1.5*IQR)
min_r,max_r
```

```python
dataset = dataset[dataset["CoapplicantIncome"]<=max_r]
```

```python
sns.distplot(dataset["CoapplicantIncome"])  # there is the change in graph
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=np.log1p)
```

```python
ft.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_tf"] = ft.transform(dataset[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(dataset["CoapplicantIncome"])
plt.title("Before")

plt.subplot(1,2,2)
sns.distplot(dataset["CoapplicantIncome_tf"])
plt.title("After")
plt.show()
```

# Linear Regression Algorithm(Simple linear)


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split



```

```python
dataset = pd.read_csv('placement.csv')
dataset.head(3)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa",y="package",data=dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x = dataset[["cgpa"]]
y = dataset[["package"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression     #linearregression is a class
```

```python
lr = LinearRegression()                  # lr is a object
lr.fit(x_train,y_train)
```

```python
lr.coef_               # to check y= m*x+c  , m ke value
```

```python
lr.intercept_              # to check c ke value
```

```python
# y = 0.57425647*x-1.02700694
```

```python
lr.score(x_test,y_test)*100     #to check accuracy
```

```python
lr.predict([[6.89]])
```

```python
 0.57425647*6.89-1.02700694
```

```python
y_prd = lr.predict(x)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="package",data=dataset)
plt.plot(dataset["cgpa"],y_prd,c="red")
plt.legend(["org data","predict line"])
plt.savefig("predict.jpg")
plt.show()
```
