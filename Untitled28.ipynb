```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 1. Generate some sample 2D data
np.random.seed(42)
X = np.random.multivariate_normal(mean=[0, 0], cov=[[10, 5], [5, 2]], size=100)

print("Original data shape:", X.shape)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], alpha=0.7)
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)

# 2. Standardize the data
# PCA is sensitive to the variances of the variables, so standardization is crucial.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Calculate the Covariance Matrix
# The covariance matrix indicates how much each variable varies from the mean
# with respect to each other.
cov_matrix = np.cov(X_scaled.T)
print("\nCovariance Matrix:\n", cov_matrix)

# 4. Calculate Eigenvalues and Eigenvectors
# Eigenvectors represent the principal components (directions of maximal variance).
# Eigenvalues represent the magnitude of variance along those directions.
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)

# 5. Sort Eigenvalues and Corresponding Eigenvectors
# Sort in descending order to identify the most significant principal components.
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

# 6. Select Principal Components (e.g., choose top 2 components for 2D data)
# For 2D data, we typically keep both components unless reducing dimensionality.
num_components = 2 # Keeping both for visualization
principal_components = sorted_eigenvectors[:, :num_components]

print("\nSelected Principal Components (Eigenvectors):\n", principal_components)

# 7. Transform the Data
# Project the standardized data onto the selected principal components.
X_pca = X_scaled.dot(principal_components)

print("\nTransformed data shape:", X_pca.shape)

# 8. Visualize the Transformed Data
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.7)
plt.title('Data Transformed by PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)

plt.tight_layout()
plt.show()
```
