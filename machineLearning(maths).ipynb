# Central Limit Theorem

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Pop_data =[np.random.randint(10,100) for i in range(10000)]
Pop_table = pd.DataFrame({"Pop_data":Pop_data})
Pop_table
```

```python
np.mean(Pop_data)
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="Pop_data",data=Pop_table)
plt.show()
```

```python
sample_mean =[]
for no_sample in range(50):
  sample_data =[]
  for data in range(500):
    sample_data.append(np.random.choice(Pop_data))
  sample_mean.append(np.mean(sample_data))
```

```python
np.mean(sample_mean)
```

```python
sample_M = pd.DataFrame({"sample_mean":sample_mean})
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="sample_mean",data=sample_M)
plt.show()
```

# Hypothesis testing
z-test
example 1 = A teacher claims that the mean score of students in his class is greater than 82 with a standard deviation of 20.if a sample of 81 students was selected with a mean score of 90





```python
  import scipy.stats as st
  import numpy as np
```

```python
s_x = 90
p_u = 82
p_std = 20
n = 81
ap = 0.05
```

```python
z_cal = (s_x-p_u)/(p_std/np.sqrt(n))
z_cal
```

```python
z_table =st.norm.ppf(1-ap)
z_table
```

```python
if z_table < z_cal :
  print ("ha is right")
else:
  print("ho is right")
```
