# Outlier Removal using IQR

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
sns.boxplot(x = "CoapplicantIncome",data = dataset)
plt.show()
```

```python
dataset.shape
```

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
min_range = q1 - (1.5 * IQR)
max_range = q3 + (1.5 * IQR)
min_range,max_range
```

```python
new_dataset = dataset[dataset["CoapplicantIncome"]<=max_range]
new_dataset
```

```python
new_dataset.shape
```

```python
sns.boxplot(x = "CoapplicantIncome",data = new_dataset)
plt.show()
```

# Outlier Removal using Z-Score

```python
 import pandas as pd
 import seaborn as sns
 import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.boxplot(x = "CoapplicantIncome",data = dataset)
```

```python
sns.displot(dataset["CoapplicantIncome"])
```

```python
min_range = dataset["CoapplicantIncome"].mean() - (3 * dataset["CoapplicantIncome"].std())
man_range = dataset["CoapplicantIncome"].mean() + (3 * dataset["CoapplicantIncome"].std())
max_range,min_range
```

```python
new_data = dataset[dataset["CoapplicantIncome"]<=max_range]
```

```python
sns.boxplot(x = "CoapplicantIncome",data = new_data)
```

```python
Z_Score = (dataset["CoapplicantIncome"] - dataset["CoapplicantIncome"].mean())/(dataset["CoapplicantIncome"].std())
Z_Score
```

```python
dataset["Z_Score"] = Z_Score
```

```python
dataset[dataset["Z_Score"]<3]
```

# Feature Scaling (Standardization)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.displot(dataset["ApplicantIncome"], kde=True)
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"] = pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=["X"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
#plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.histplot(dataset["ApplicantIncome_ss"] , kde=True)

plt.subplot(1,2,2)
plt.title("After")
sns.histplot(dataset["ApplicantIncome"], kde=True)

plt.show()
```

# Feature Scaling(Normalization)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_min"] = ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.histplot(dataset["CoapplicantIncome"], kde=True)

plt.subplot(1,2,2)
plt.title("After")
sns.histplot(dataset["CoapplicantIncome_min"], kde=True)

plt.tight_layout()
plt.show()
```
